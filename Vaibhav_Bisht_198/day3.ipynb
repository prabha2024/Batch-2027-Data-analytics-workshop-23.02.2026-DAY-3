```python
import statistics

numbers = [10, 20, 30, 40, 50, 20]
mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)
mode_value = statistics.mode(numbers)
variance_value = statistics.variance(numbers)
std_deviation = statistics.stdev(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Variance:", variance_value)
print("Standard Deviation:", std_deviation)
```

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])
print("Addition:\n", A + B)
print("Subtraction:\n", A - B)
print("Multiplication:\n", A @ B)
print("Transpose of A:\n", A.T)
```

```python

import numpy as np

arr = np.random.randint(0, 101, 10)
print("Array:", arr)
print("Mean:", arr.mean())
print("Standard Deviation:", arr.std())
print("Minimum:", arr.min())
print("Maximum:", arr.max())
```

```python
import numpy as np
from scipy import stats
data = np.array([10, 12, 9, 11, 10, 13, 12, 8])
mean_value = stats.tmean(data)
print("Mean:", mean_value)
t_stat, p_value = stats.ttest_1samp(data, 10)
print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python

from scipy import integrate
import numpy as np
def f(x):
    return x**2
result, error = integrate.quad(f, 0, 3)

print("Definite Integration Result:", result)
print("Estimated Error:", error)
```

```python

from scipy import optimize
def f(x):
    return x**2 + 4*x + 4
result = optimize.minimize(f, x0=0)
print("Minimum Value:", result.fun)
print("At x =", result.x)
```
