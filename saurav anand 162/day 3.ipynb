1. Write a Python program to calculate the following for a given list of numbers:
•	Mean
•	Median
•	Mode
•	Variance
•	Standard Deviation


```python
import statistics
import math

numbers = [10,20,30,40,50,20]

mean = sum(numbers)/ len(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)
variance = statistics.variance(numbers)
std_dev = statistics.stdev(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

Create a NumPy array and perform the following operations:
•	Find sum and mean
•	Find maximum and minimum
•	Reshape the array
•	Perform element-wise multiplication


```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

# Sum and Mean
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))

# Max and Min
print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))

# Reshape
reshaped = arr.reshape(2, 3)
print("Reshaped Array:\n", reshaped)

# Element-wise multiplication
print("Element-wise multiplication:\n", reshaped * 2)
```

. Write a Python program to:
•	Create two matrices
•	Perform addition and subtraction
•	Perform matrix multiplication
•	Find transpose of a matrix


```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Addition
print("Addition:\n", A + B)

# Subtraction
print("Subtraction:\n", A - B)

# Matrix Multiplication
print("Multiplication:\n", np.dot(A, B))

# Transpose
print("Transpose of A:\n", A.T)
```

4. Generate a random array of 10 numbers and calculate:
•	Mean
•	Standard deviation
•	Minimum and Maximum value


```python
import numpy as np

arr = np.random.rand(10)  # 10 random numbers

print("Array:", arr)
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))
print("Minimum:", np.min(arr))
print("Maximum:", np.max(arr))
```

5. Write a Python program using SciPy to:
•	Calculate mean using scipy.stats
•	Perform a t-test on a sample dataset


```python
from scipy import stats
import numpy as np

data = np.array([10, 12, 14, 16, 18])

# Mean using scipy
mean = stats.tmean(data)
print("Mean:", mean)

# One sample t-test (test mean = 12)
t_stat, p_value = stats.ttest_1samp(data, 12)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

6. Write a Python program to compute definite integration of a mathematical function using scipy.integrate.

```python
from scipy import integrate
import numpy as np

# Define function
def f(x):
    return x**2

# Integrate from 0 to 5
result, error = integrate.quad(f, 0, 5)

print("Integration Result:", result)
```

7. Write a Python program to minimize a mathematical function using scipy.optimize.

```python
from scipy import optimize

# Function to minimize
def f(x):
    return x**2 + 4*x + 4

result = optimize.minimize(f, 0)

print("Minimum Value:", result.fun)
print("At x =", result.x)
```
